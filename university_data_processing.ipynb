```python
import os
import re
import numpy as np
import pandas as pd
from PIL import Image
from tqdm import tqdm
from ast import literal_eval
import h5py
```

```python
# Define the directory containing the image files
images_directory = 'data/part10'

# Define a list to hold the data for each image
data = []

# Loop over each file in the directory
for filename in tqdm(os.listdir(images_directory)):
    # Extract the placemark and view from the filename
    match = re.match(r'(\d{6})_(\d)\.jpg', filename)
    if match:
        placemark = match.group(1)
        view = int(match.group(2))
        
        # Load the image
        filepath = os.path.join(images_directory, filename)
        image = Image.open(filepath)
        
        # Convert the image to a numpy array
        image = np.array(image)
        
        # Append the data to the list
        data.append({'filename': filename, 'placemark': placemark, 'view': view, 'image': image})

# Create a dataframe from the data
df = pd.DataFrame(data)
```

```python
# Make a copy of the df dataframe
df_copy = df.copy()
```

```python
# Define the new size for the images
new_size = (200, 250)

# Loop over each row in the dataframe
for index, row in df_copy.iterrows():
    # Load the image as a PIL Image object
    image = Image.fromarray(row['image'])
    
    # Resize the image
    resized_image = image.resize(new_size)
    
    # Update the image data in the dataframe
    df_copy.at[index, 'image'] = np.array(resized_image)
```
